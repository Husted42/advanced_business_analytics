```python
import pandas as pd

data_path_trump_features = "../data/trump_daily_features.csv"
data_path_news_features = "../Newsdata/sentiment_2025-11-14_to_2026-04-10.csv"
data_path_exogenous_features = "../Inflation/merged_macro_daily_2025_2026.csv"
```

```python
# load data into pandas dataframes
pd_trump_features = pd.read_csv(data_path_trump_features)
pd_trump_features["date"] = pd.to_datetime(pd_trump_features["post_date"])

pd_news_features = pd.read_csv(data_path_news_features)
pd_news_features["date"] = pd.to_datetime(pd_news_features["post_date"])

pd_news_macro = pd.read_csv(data_path_exogenous_features)
pd_news_macro["date"] = pd.to_datetime(pd_news_macro["date"])
```

```python
print("Trump columns")
print(pd_trump_features.columns)
```

```python
# merge dataframes on date
def merge_dataframes_on_date(df1, df2):
    merged_df = pd.merge(df1, df2, on="date", how="inner")
    return merged_df

df_full = merge_dataframes_on_date(pd_trump_features, pd_news_features)
df_full = merge_dataframes_on_date(df_full, pd_news_macro)
print(df_full.columns)
```
